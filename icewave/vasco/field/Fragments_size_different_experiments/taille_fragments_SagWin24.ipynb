```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
facq_t = 30
wavelength = 19.5
wavelength_err = 1.5

facq_x_approx = 1200/100 # px/m


path2resultsimagj = "C:/Users/Vasco Zanchi/Desktop/Saguenay2024/0211/Drones/Results/tailles_fragments/Results_fragmentsizes.csv"

df = pd.read_csv(path2resultsimagj)

Sizes_rawdata = df['Length'].values

array_L_fragments = Sizes_rawdata[::2]
array_w_fragments = Sizes_rawdata[1::2]

array_L_fragments = array_L_fragments * (1/facq_x_approx)
array_w_fragments = array_w_fragments * (1/facq_x_approx)

print(array_L_fragments)
print(array_w_fragments)


path2frames_fragments = "C:/Users/Vasco Zanchi/Desktop/Saguenay2024/0211/Drones/Results/tailles_fragments/frames_fragments.csv"
df2 = pd.read_csv(path2frames_fragments, encoding='latin1', sep=';')

array_frames = np.int64(df2['frame'].values)
print(array_frames)
```

```python
plt.figure()
plt.plot(array_L_fragments,array_w_fragments,'o')
plt.xlabel('Length of the fragments [m]')
plt.ylabel('Width of the fragments [m]')
```

```python
x2plot = array_frames
y2plot = array_L_fragments/array_w_fragments
plt.plot(array_frames * 1/facq_t, y2plot,'o')
plt.ylim(0,np.max(y2plot)*1.1)
plt.xlabel('Time [seconds]')
plt.ylabel('Length/width of the fragments')
```

```python
x2plot = array_frames
y2plot = array_w_fragments
plt.plot(array_frames * 1/facq_t, y2plot,'o')
plt.ylim(0,np.max(y2plot)*1.1)
plt.xlabel('Time [seconds]')
plt.ylabel('Length/width of the fragments')
```

```python
x2plot = array_frames
y2plot = array_L_fragments
plt.plot(array_frames * 1/facq_t, y2plot,'o')
plt.ylim(0,np.max(y2plot)*1.1)
plt.xlabel('Time [seconds]')
plt.ylabel('Length of the fragments  [m]')
```
