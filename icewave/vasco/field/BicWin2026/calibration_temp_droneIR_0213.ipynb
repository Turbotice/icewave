```python
# dans cette partie conversion de video mp4 en images tiff
import cv2
import os
import numpy as np
from PIL import Image
def video_to_images(video_path, output_folder):
    # Créer le dossier de sortie s'il n'existe pas
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Ouvrir la vidéo
    cap = cv2.VideoCapture(video_path)
    frame_count = 0

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        # Convertir l'image de BGR (OpenCV) à RGB (PIL)
        rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

        # Convertir en image PIL et sauvegarder en TIFF
        pil_image = Image.fromarray(rgb_frame)
        output_path = os.path.join(output_folder, f"frame_{frame_count:04d}.tiff")
        pil_image.save(output_path)

        frame_count += 1

    cap.release()
    print(f"Conversion terminée. {frame_count} images sauvegardées dans '{output_folder}'.")
# Exemple d'utilisation
video_path = r"F:\bicwin26\data\0213\Drones\Tourterelle\videos_calib_temp\DJI_20260213192522_0005_S.MP4"
output_folder = r"F:\bicwin26\data\0213\Drones\Tourterelle\images_calib_temp\DJI_20260213192522_0005_S"
video_to_images(video_path, output_folder)
```
